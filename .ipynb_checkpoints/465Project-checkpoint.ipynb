```python
# X_new, y_new = [], []
# max_total = 3000
# X_0, X_1, X_2 = X[y == 0], X[y == 1], X[y == 2]

# for label, X_set in zip(label_map.keys(), [X_0, X_1, X_2]):
#     count = 0
#     for img in X_set:
#         if count > max_total - label_counts[label]:
#             break
#         for i in range(2):
#             augmented_img = data_augmentation(img)
#             X_new.append(augmented_img)
#             y_new.append(label)
#             count += 1

# X_new, y_new = np.array(X_new), np.array(y_new)
# print(X_new.shape, y_new.shape)

import tensorflow as tf
import numpy as np

# Define the improved data augmentation pipeline
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),              # Horizontal flip
    tf.keras.layers.RandomRotation(0.05),                  # Small rotation (5%)
    tf.keras.layers.RandomTranslation(0.05, 0.05),         # Small shift (5%)
    tf.keras.layers.RandomContrast(0.15),                  # Slight contrast variation (15%)
    tf.keras.layers.Lambda(lambda x: x + tf.random.normal(tf.shape(x), mean=0.0, stddev=0.01)),  # Add Gaussian noise
])

# Initialize lists for new augmented data
X_new, y_new = [], []
max_total = 3000
X_0, X_1, X_2 = X[y == 0], X[y == 1], X[y == 2]

# Generate augmented data per class to balance
for label, X_set in zip(label_map.keys(), [X_0, X_1, X_2]):
    count = 0
    for img in X_set:
        if count >= max_total - label_counts[label]:
            break
        for _ in range(3):  # Generate three augmentations per image
            augmented_img = data_augmentation(img[None, ...])  # Add batch dimension
            X_new.append(tf.squeeze(augmented_img).numpy())    # Squeeze and convert to numpy
            y_new.append(label)
            count += 1

# Convert lists to arrays and ensure dimensions match
X_new, y_new = np.array(X_new), np.array(y_new)

# Ensure `X` and `X_new` have the same number of dimensions
if X.ndim == 3:  # If X is [num_samples, height, width]
    X = np.expand_dims(X, axis=-1)  # Convert to [num_samples, height, width, channels]
if X_new.ndim == 3:
    X_new = np.expand_dims(X_new, axis=-1)

# Concatenate original and augmented datasets
X = np.concatenate([X, X_new], axis=0)
y = np.concatenate([y, y_new], axis=0)

print("Final augmented dataset shape:", X.shape, y.shape)
```

```python
import os

import keras
from keras import layers
from keras import ops
import numpy as np
```

```python
class Distiller(keras.Model):
    def __init__(self, student, teacher):
        super().__init__()
        self.teacher = teacher
        self.student = student

    def compile(
        self,
        optimizer,
        metrics,
        student_loss_fn,
        distillation_loss_fn,
        alpha=0.5,
        temperature=3,
    ):
        """Configure the distiller.

        Args:
            optimizer: Keras optimizer for the student weights
            metrics: Keras metrics for evaluation
            student_loss_fn: Loss function of difference between student
                predictions and ground-truth
            distillation_loss_fn: Loss function of difference between soft
                student predictions and soft teacher predictions
            alpha: weight to student_loss_fn and 1-alpha to distillation_loss_fn
            temperature: Temperature for softening probability distributions.
                Larger temperature gives softer distributions.
        """
        super().compile(optimizer=optimizer, metrics=metrics)
        self.student_loss_fn = student_loss_fn
        self.distillation_loss_fn = distillation_loss_fn
        self.alpha = alpha
        self.temperature = temperature

    def compute_loss(
        self, x=None, y=None, y_pred=None, sample_weight=None, allow_empty=False
    ):
        teacher_pred = self.teacher(x, training=False)
        student_loss = self.student_loss_fn(y, y_pred)

        distillation_loss = self.distillation_loss_fn(
            ops.softmax(teacher_pred / self.temperature, axis=1),
            ops.softmax(y_pred / self.temperature, axis=1),
        ) * (self.temperature**2)

        loss = self.alpha * student_loss + (1 - self.alpha) * distillation_loss
        return loss

    def call(self, x):
        return self.student(x)
```

```python
import tensorflow as tf
import tensorflow_addons as tfa

# Define losses
categorical_crossentropy = tf.keras.losses.CategoricalCrossentropy()
mse_loss = tf.keras.losses.MeanSquaredError()
contrastive_loss = tfa.losses.ContrastiveLoss()

# Functions for different components
def soft_target_loss(teacher_logits, student_logits, temperature=3.0):
    teacher_probs = tf.nn.softmax(teacher_logits / temperature)
    student_probs = tf.nn.softmax(student_logits / temperature)
    return tf.reduce_mean(-tf.reduce_sum(teacher_probs * tf.math.log(student_probs), axis=1))

def feature_loss(teacher_features, student_features):
    return mse_loss(teacher_features, student_features)

def attention_loss(teacher_features, student_features):
    def compute_attention_maps(features):
        return tf.reduce_mean(features, axis=-1)  # Simple channel-wise mean
    teacher_attention = compute_attention_maps(teacher_features)
    student_attention = compute_attention_maps(student_features)
    return mse_loss(teacher_attention, student_attention)

def contrastive_knowledge_loss(student_embeddings, teacher_embeddings, labels):
    positive_pairs = tf.linalg.matmul(labels, labels, transpose_b=True)
    return contrastive_loss(teacher_embeddings, student_embeddings, positive_pairs)

# Extract intermediate outputs
def get_intermediate_output(model, layer_name, input_data):
    intermediate_layer_model = tf.keras.Model(inputs=model.input, outputs=model.get_layer(layer_name).output)
    return intermediate_layer_model(input_data)

# Unified training loop
def train_student(student, teacher, train_data, layers, temperature=3.0):
    optimizer = tf.keras.optimizers.Adam()
    
    for x, y in train_data:
        with tf.GradientTape() as tape:
            # Forward pass
            student_logits = student(x, training=True)
            teacher_logits = teacher(x, training=False)

            # 1. Response-Based Distillation Loss
            soft_loss = soft_target_loss(teacher_logits, student_logits, temperature)

            # 2. Feature-Based and Attention-Based Loss
            total_feature_loss, total_attention_loss = 0, 0
            for layer_name in layers:
                teacher_features = get_intermediate_output(teacher, layer_name, x)
                student_features = get_intermediate_output(student, layer_name, x)
                total_feature_loss += feature_loss(teacher_features, student_features)
                total_attention_loss += attention_loss(teacher_features, student_features)

            # 3. Contrastive Loss
            teacher_embeddings = teacher(x, training=False)  # Assume logits can represent embeddings
            student_embeddings = student(x, training=False)
            contrastive_loss_value = contrastive_knowledge_loss(student_embeddings, teacher_embeddings, y)

            # Combine all losses
            total_loss = soft_loss + total_feature_loss + total_attention_loss + contrastive_loss_value

        # Backward pass
        gradients = tape.gradient(total_loss, student.trainable_variables)
        optimizer.apply_gradients(zip(gradients, student.trainable_variables))

        # Logging (optional)
        print(f"Loss: {total_loss.numpy():.4f} | Soft: {soft_loss.numpy():.4f}, Feature: {total_feature_loss.numpy():.4f}, "
              f"Attention: {total_attention_loss.numpy():.4f}, Contrastive: {contrastive_loss_value.numpy():.4f}")
```

```python
import tensorflow as tf
from tensorflow.keras.losses import MeanSquaredError
import tensorflow_addons as tfa

# Define individual loss functions
mse_loss = MeanSquaredError()
contrastive_loss = tfa.losses.ContrastiveLoss()

def compute_attention_maps(features):
    return tf.reduce_mean(features, axis=-1)  # Simple channel-wise mean

def distillation_loss(teacher_logits, student_logits, temperature):
    teacher_probs = tf.nn.softmax(teacher_logits / temperature)
    student_probs = tf.nn.softmax(student_logits / temperature)
    return tf.reduce_mean(-tf.reduce_sum(teacher_probs * tf.math.log(student_probs), axis=1))

def compute_crd_loss(student_embeddings, teacher_embeddings, labels):
    positive_pairs = tf.linalg.matmul(labels, labels, transpose_b=True)
    return contrastive_loss(teacher_embeddings, student_embeddings, positive_pairs)

# Training loop
def train_student(teacher, student, data, layer_name, temperature, 
                  lambda_response=0.5, lambda_feature=0.2, lambda_attention=0.2, lambda_contrastive=0.1):
    optimizer = tf.keras.optimizers.Adam()
    
    for epoch in range(10):  # Number of epochs
        epoch_loss = 0
        for x, y in data:
            with tf.GradientTape() as tape:
                # Forward passes
                teacher_logits = teacher(x, training=False)
                student_logits = student(x, training=True)
                
                # Response-based distillation loss
                response_loss = distillation_loss(teacher_logits, student_logits, temperature)
                
                # Feature-based distillation loss
                teacher_features = get_intermediate_output(teacher, layer_name, x)
                student_features = get_intermediate_output(student, layer_name, x)
                feature_loss = mse_loss(teacher_features, student_features)
                
                # Attention-based distillation loss
                teacher_attention = compute_attention_maps(teacher_features)
                student_attention = compute_attention_maps(student_features)
                attention_loss = mse_loss(teacher_attention, student_attention)
                
                # Contrastive distillation loss
                crd_loss = compute_crd_loss(student_features, teacher_features, y)
                
                # Combine losses
                total_loss = (
                    lambda_response * response_loss +
                    lambda_feature * feature_loss +
                    lambda_attention * attention_loss +
                    lambda_contrastive * crd_loss
                )
                
            # Backpropagation
            grads = tape.gradient(total_loss, student.trainable_variables)
            optimizer.apply_gradients(zip(grads, student.trainable_variables))
            
            epoch_loss += total_loss.numpy()
        
        print(f"Epoch {epoch+1}: Loss = {epoch_loss / len(data)}")
```
